```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
star_wars_df = pd.read_csv(r"D:\exel\Protfolio Project\Python project\Star wars surver data cleaning project\StarWars.csv", encoding="ISO-8859-1")
```

```python
#Let's replace the str type values 'Yes' and 'No' in columns: 1-2 to Boolean values
```

```python
star_wars_df['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
star_wars_df['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
Type_transfer_dic = {"Yes": True,
                     "No": False}

for col in ['Have you seen any of the 6 films in the Star Wars franchise?',
            'Do you consider yourself to be a fan of the Star Wars film franchise?']:
    
    star_wars_df[col] = star_wars_df[col].map(Type_transfer_dic)
```

```python
star_wars_df['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
star_wars_df['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
#In the columns: "Which of the following Star Wars films have you seen? Please select all that apply." till column "Unnamed: 8"
#the participants are asked to select the film if they saw it(1 column for each film).
#I will change the columns name so they will indicate the film and change the values to True and False.
```

```python
star_wars_df.iloc[: ,3:9]
```

```python
star_wars_df = star_wars_df.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

```

```python
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.NaN: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}


for col in ['seen_1', 'seen_2', 'seen_3', 'seen_4', 'seen_5', 'seen_6']:
    
    star_wars_df[col] = star_wars_df[col].map(movie_mapping)
```

```python
star_wars_df.iloc[: ,3:9]
```

```python
# in the columns 9:5 the participants are asked to rank their favorite Star Wars episode (rank place for each episode)
# I will change the columns name to "Episode 1 Ranking", "Episode 2 Ranking".... and the dtype to float,
# so I will be able to visualise the results
```

```python
star_wars_df[star_wars_df.columns[9:15]]
```

```python
# removing the first row and changing data type
star_wars_df = star_wars_df.iloc[1: , :]
```

```python
star_wars_df[star_wars_df.columns[9:15]] = star_wars_df[star_wars_df.columns[9:15]].astype(float)
```

```python
star_wars_df.dtypes
```

```python
#change columns name
star_wars_df = star_wars_df.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "Episode 1 Ranking",
        "Unnamed: 10": "Episode 2 Ranking",
        "Unnamed: 11": "Episode 3 Ranking",
        "Unnamed: 12": "Episode 4 Ranking",
        "Unnamed: 13": "Episode 5 Ranking",
        "Unnamed: 14": "Episode 6 Ranking"
        })
```

```python
# showing the episode ranks
star_wars_df[star_wars_df.columns[9:15]].mean()
```

```python
plt.figure(figsize = (6,4))
plt.bar(['Episode 1', 'Episode 2', 'Episode 3', 'Episode 4', 'Episode 5', 'Episode 6'], star_wars_df[star_wars_df.columns[9:15]].mean(), color = 'blue', width = 0.4)
plt.title('Average Ranking of each Star Wars Movie(lower bar means better rank)')
plt.xlabel('Star war episode')
plt.ylabel('Rank');
```

```python
#showing the sum of views for each movie and the corraletion between the rank of movie and the amount of people who watched the movie 
```

```python
star_wars_df[star_wars_df.columns[3:9]].sum()
```

```python
plt.figure(figsize = (6,4))
plt.bar(range(1,7), star_wars_df[star_wars_df.columns[3:9]].sum(), color = 'blue', width = 0.4)
plt.title('Sum of views for each movie')
plt.xlabel('Star Wars episode')
plt.ylabel('Sum of views');
```

```python
# As we can see there is a clear corraletion between the rank and the sum of views.
# For instance episode 5 is the most watched and has the highest rank
```

```python
# diffrence between male and female fans
```

```python
male_fans = star_wars_df[star_wars_df["Gender"] == "Male"]
female_fans = star_wars_df[star_wars_df["Gender"] == "Female"]
```

```python

plt.figure(figsize = (8,4))
plt.bar(range(1,7), male_fans[male_fans.columns[9:15]].mean(), align = 'edge', width = -0.4)
plt.bar(range(1,7), female_fans[female_fans.columns[9:15]].mean(),  align = 'edge', width = 0.4)
plt.title('AVG rank by male fans VS  female fans(lower bar means better rank)')
plt.xlabel('Star Wars episode')
plt.ylabel('Rank')
plt.legend(['male_fans', 'female_fans']);
```

```python
plt.figure(figsize = (8,4))
plt.bar(range(1,7), male_fans[male_fans.columns[3:9]].sum(), align = 'edge', width = -0.4)
plt.bar(range(1,7), female_fans[female_fans.columns[3:9]].sum(),  align = 'edge', width = 0.4)
plt.title('Sum of views for each movie by Gender')
plt.xlabel('Star Wars episode')
plt.ylabel('Views amount')
plt.legend(['male_fans', 'female_fans']);
```

```python
# Showing the corraletion between Star Wars and Star Trek fans
```

```python
Star_wars_fans_amount = len( star_wars_df.loc[star_wars_df['Do you consider yourself to be a fan of the Star Wars film franchise?'] == True])
Star_trek_fans_amount = len( star_wars_df.loc[star_wars_df['Do you consider yourself to be a fan of the Star Trek franchise?'] == "Yes"])




plt.figure(figsize = (4,4))
plt.bar(['Star wars fans', 'Star Trek fans'], [Star_wars_fans_amount, Star_trek_fans_amount], align = 'edge', width = 0.4)
plt.title('Corraletion between Star Wars fans and Star Trek fans')
plt.ylabel('Fans amount');
```

```python
# There is corraletion!  most of the Star Wars fans were also Star Trek fans
```
