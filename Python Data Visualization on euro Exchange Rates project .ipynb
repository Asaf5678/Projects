```python
#importing the libraries and the data I am going to use
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


exchange_rates = pd.read_csv(r"D:\exel\Protfolio Project\Python project\Euro exchange rates data visialisation project\euro-daily-hist_1999_2022.csv")
```

```python
#rename columns and changing Dtype for the ['Time'] column
```

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
```

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
#Showing corraletion between diffrent currencies In with regard to the Euro
```

```python
exchange_rates_corr = exchange_rates.iloc[: , 1: ]
```

```python
for col in exchange_rates_corr.columns:
    exchange_rates_corr.loc[exchange_rates_corr[col] == '-', col] = None
    exchange_rates_corr[col] = exchange_rates_corr[col].astype(float)
```

```python
exchange_rates_corr.corr(method ='pearson')
```

```python
plt.figure(figsize = (14, 12))
correlation_matrix = exchange_rates_corr.iloc[ : ,11 :  22].corr(method ='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for different currencies Euro exchange rate", weight='bold')

plt.xlabel("currency name")

plt.ylabel("currency name")

plt.show()
```

```python
# As we can see there is a high correlation between the Indonesian rupiah snd the Indian rupee euro exchange rate
```

```python
# Creating a table for only the Euro to US dollar Exchange rate and changing us dollar Dtype to float
```

```python
euro_to_dollar = exchange_rates.iloc[ : , [0, 39]]
```

```python
euro_to_dollar.head(5)
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar.loc[ euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

```python
# showing the Euro to US dollar exchange rate over the time 
```

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'].rolling(30).mean())
plt.title('Euro to dollar exchange rate over the time with a rolling window of a month', weight='bold')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout()
plt.show()
```

```python
# Euro to dollar exchange rate during Bush, Obama and Trump presidensy 
```

```python
euro_to_dollar_Bush = euro_to_dollar.loc[(euro_to_dollar['Time'] > '1999-12-31') &
                                         (euro_to_dollar['Time'] <= '2008-12-31')]


euro_to_dollar_Obama = euro_to_dollar.loc[(euro_to_dollar['Time'] > '2008-12-31') &
                                         (euro_to_dollar['Time'] <= '2016-12-31')]


euro_to_dollar_Trump = euro_to_dollar.loc[(euro_to_dollar['Time'] > '2016-12-31') &
                                         (euro_to_dollar['Time'] <= '2020-12-31')]
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(11,7))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar_Bush['Time'], euro_to_dollar_Bush['US_dollar'].rolling(30).mean())
plt.title('Bush', weight='bold')
plt.ylabel('Euro to US dollar')

plt.subplot(3,2,2)
plt.plot(euro_to_dollar_Obama['Time'], euro_to_dollar_Obama['US_dollar'].rolling(30).mean())
plt.title('Obama ', weight='bold')
plt.ylabel('Euro to US dollar')

plt.subplot(3,2,3)
plt.plot(euro_to_dollar_Trump['Time'], euro_to_dollar_Trump['US_dollar'].rolling(30).mean())
plt.title('Trump ', weight='bold')
plt.ylabel('Euro to US dollar')

    
plt.tight_layout()
plt.show()
```

```python
### Ax4: Bush-Obama-Trump\
plt.figure(figsize=(25,10))
plt.plot(euro_to_dollar_Bush['Time'], euro_to_dollar_Bush['US_dollar'].rolling(30).mean(),
        color='purple', linewidth=9)
plt.plot(euro_to_dollar_Obama['Time'], euro_to_dollar_Obama['US_dollar'].rolling(30).mean(),
        color='orange', linewidth=9)
plt.plot(euro_to_dollar_Trump['Time'], euro_to_dollar_Trump['US_dollar'].rolling(30).mean(),
        color='white', linewidth=9)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(['Bush', 'Obama', 'Trump'], fontsize="30");
```
